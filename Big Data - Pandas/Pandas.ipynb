```python
#GroupBy Operation

data1 = { 'Name' : ['Jay','Anuj','Jay','Prince','Gaurav','Anuj','Prince','Abhi'],
          'Age' : [27,24,22,32,33,26,27,32],
          'Address' : ['Nagpur','kanpur','Allahbad','Pune','Mumbai','Nagpur','Kanpur','Pune'],
          'Qualification': ['Msc','Btech','Msc','MCA','Phd','Btech','MCA','Phd']
}
```

```python
type(data1)
```

```python
df_new = pd.DataFrame(data1)
```

```python
df_new
```

```python
#Group by using single key
df_new.groupby('Name').groups
```

```python
gk = df_new.groupby('Name')
for i in gk:
    print(i)
```

```python
#GroupBy using multipke keys

data1 = { 'Name' : ['Jay','Anuj','Jay','Prince','Gaurav','Anuj','Prince','Abhi'],
          'Age' : [27,24,22,32,33,26,27,32],
          'Address' : ['Nagpur','kanpur','Allahbad','Pune','Mumbai','Nagpur','Kanpur','Pune'],
          'Qualification': ['Msc','Btech','Msc','MCA','Phd','Btech','MCA','Phd']
}
df_new = pd.DataFrame(data1)
df_new
```

```python
df_new.groupby(['Name','Qualification']).groups
```

```python
#Sorting the values based on column - by defualt it will be ascending order
df_new.sort_values('Age')
```

```python
list(df_new.groupby(['Address']))
```

```python

#Sort in descending order
df_new.sort_values('Age',ascending=0)
```

```python
#Merger two dataframes

data1 = { 'Name' : ['Jay','Anuj','Jay','Prince'],
          'Age' : [27,24,22,32],
          'Address' : ['Nagpur','kanpur','Allahbad','Pune'],
          'Qualification': ['Msc','Btech','Msc','MCA']
}

data2= { 'Name' : ['Gaurav','Anuj','Prince','Abhi'],
          'Age' : [33,26,27,32],
          'Address' : ['Mumbai','Nagpur','Kanpur','Pune'],
          'Qualification': ['Phd','Btech','MCA','Phd']
}

df1 = pd.DataFrame(data1,index =[0,1,2,3])
df2 = pd.DataFrame(data2, index = [4,5,6,7])

df_concat=[df1,df2]
result = pd.concat(df_concat)
result
```
